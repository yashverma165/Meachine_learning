# Decision tree Regressor

```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
```

```python
from sklearn.datasets import fetch_california_housing
california_df=fetch_california_housing()
```

```python
california_df
```

```python
# Independent Feature
x=pd.DataFrame(california_df.data,columns=california_df.feature_names)
# dependent Feature
y=california_df.target
```

```python
x.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=.33)
```

```python
regressor=DecisionTreeRegressor()
regressor.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
y_pred=regressor.predict(x_test)
print(y_pred)
print(r2_score(y_pred,y_test))
```

## Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV
parameters={'criterion':['squared_error','friedman_mse', 'absolute_error', 'poisson'],
           'splitter':['best', 'random'],
            'max_depth':[1,2,3,4,5,6],
          'max_features':['auto', 'sqrt', 'log2']}
```

```python
regcv=DecisionTreeRegressor()
```

```python
regcv=GridSearchCV(regcv,param_grid=parameters,cv=5,scoring='neg_mean_squared_error')
```

```python
regcv.fit(x_train,y_train)
```

```python
regcv.best_params_
```

```python
regcv=DecisionTreeRegressor( criterion = 'poisson',
                             max_depth= 6,
                             max_features= 'auto',
                             splitter = 'best')
```

```python
regcv.fit(x_train,y_train)
y_pred=regcv.predict(x_test)
y_pred
```

```python
print(r2_score(y_pred,y_test))
```
